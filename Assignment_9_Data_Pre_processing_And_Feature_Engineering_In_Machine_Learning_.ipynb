**Discuss the pros and cons of One-Hot Encoding and Label Encoding.**

One-Hot Encoding (OHE) : Transforms a categorical variable into multiple binary columns (0/1), one for each unique category.

Pros: No ordinal assumption, Works well with linear models

cons: High dimensionality, Multicollinearity

Label Encoding : Assigns each category a unique integer

pros: Space-efficient, Fast and simple

cons: Imposes artificial ordering, Can mislead linear models.

**Feature Engineering:**

	Create at least 2 new features that could be beneficial for the model. Explain the rationale behind your choices

```python
# 1.Net Capital
df["net_capital"] = df["capital_gain"] - df["capital_loss"]
```

Explanation : Instead of treating capital_gain and capital_loss separately, combining them into a single net capital feature provides a clearer signal of a persons financial activity.

```python
# 2.Work Intensity
df["work_intensity"] = df["hours_per_week"] * df["education_num"]
```

Explanation : hours_per_week alone shows effort, while education_num reflects qualification level. Multiplying them captures the idea of educated effort: someone working long hours with higher education is more likely to earn >50K.
